# TASK  - Image to pencil sketch with python

# import library

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

# Read Image

```python
image=cv2.imread("flower.png")
```

# To show or display Image

```python
cv2.imshow("Flower",image)
```

# Read the image in RGB Format

```python
import matplotlib.pyplot as plt
%matplotlib inline
img_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

# Creating new image by converting the original image to grayscale

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray_image
```

```python
plt.imshow(gray_image)
```

# Inverting Image

```python
inverted_image=255 - gray_image
plt.imshow(inverted_image)
```

# Blurred the image by using Guassian Function

```python
blurred=cv2.GaussianBlur(inverted_image ,(21,21),0)
plt.imshow(blurred)
```

# Invert the Blurred image to final Image

```python
inverted_blurred=255-blurred
pencil_sketch=cv2.divide(gray_image,inverted_blurred,scale=256.0)
plt.imshow(pencil_sketch)
```
